## Sélection des premiers vols

```python
selection = []
id_vol = 0

for vol in df.iterator(100): # on sélectionne les 100 premiers vols
    vol.insert(0, 'id_vol', df.sigpos)
    selection.append(vol)

selection_df = pd.concat(selection)
selection_df
```

## Choix des variables pour la forêt aléatoire

```python
var = ['ALT', 'M [MACH]', 'TAT [DEG C]',
       'P0_1 [PSIA]', 'P0_2 [PSIA]']
# on pourra rajouter ces variables 'TLA_1 [DEG]' et 'TLA_2 [DEG]'

but = ['N1_1 [% RPM]', 'N1_2 [% RPM]',
       'N2_1 [% RPM]', 'N2_2 [% RPM]',
       'PS3_1 [PSIA]', 'PS3_2 [PSIA]',
       'T_OIL_1 [DEG C]', 'T_OIL_2 [DEG C]',
       'Q_1 [LB/H]', 'Q_2 [LB/H]']
```

```python
selection_df[but]
```

## Modèle de forêt aléatoire

```python
model = RandomForestRegressor(n_estimators = 100, max_depth = 15)
model.fit(selection_df[var], selection_df[but])
```

## Prédictions des variables

```python
residual = []
prediction = []

for i in range(len(df)):
    
    vol = df[i]
    
    pred = model.predict(vol[var])
    df_pred = pd.DataFrame(pred, columns=but, index=df.df.index)

    prediction.append(df_pred)
    
    tmp = vol[but] - df_pred
    tmp.insert(0, 'id_vol', df.sigpos)

    residual.append(tmp)
```

```python
err = pd.concat(residual)
prediction = pd.concat(prediction)
```

### Normalisation des erreurs

```python
col = list(err.columns)
col.pop(0)

err_norm = (err[col] - err[col].mean()) / err[col].std()
err_norm.insert(0, 'id_vol', err['id_vol'])
```
