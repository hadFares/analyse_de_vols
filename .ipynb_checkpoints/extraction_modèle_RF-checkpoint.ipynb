```python
import numpy as np # operations mathematiques
import numpy.random as npr # simulation d'alea
import matplotlib.pyplot as plt # plot
from sklearn.ensemble import RandomForestRegressor 
import math 
import pandas as pd
import os
```

```python
# import de tabata
import sys
from pathlib import Path

try:
    project_root = Path(__file__).resolve().parent
except NameError:
    # Cas REPL / notebook
    project_root = Path.cwd()

external_path = project_root / "external"

if str(external_path) not in sys.path:
    sys.path.insert(0, str(external_path))

import tabata as tbt
from tabata import Opset
```

```python
# On va chercher les fichiers de données
parent = Path.cwd().parent

storename = parent / "aircraft_1_decollage.h5"
ds_decollage = tbt.Opset(storename)

storename = parent / "aircraft_1_croisiere.h5"
ds_croisiere = tbt.Opset(storename)
```

```python
# On enleve supprime les fichiers extract_croisiere et extract_decollage s'ils existent deja dans le dossier
for f in ["extract_croisiere.h5","extract_decollage.h5"]:
    if os.path.exists(f):
        os.remove(f)


# On crée des fichiers avec les 100 premiers vols de  nos dataset
ds_clean_croisiere = tbt.Opset("extract_croisiere.h5")
ds_clean_decollage = tbt.Opset("extract_decollage.h5")

for i in np.arange(100):
        df = ds_croisiere[i]
        ds_clean_croisiere.put(df)
        df = ds_decollage[i]
        ds_clean_decollage.put(df)
  
```

```python
ds_clean_decollage.plot()
```

```python
df.columns.tolist()
```

```python
var = ["ALT", "M [MACH]","N1_1 [% RPM]","N2_1 [% RPM]","TAT [DEG C]"]
```

```python
# Création du dataset d'entrainement
ds_2d = pd.concat([vol for vol in ds_clean_decollage])
X_train =  ds_2d[var]
Y_train = ds_2d

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)

# Étape 3 : On lui demande de "reconstruire" les données
predictions = model.predict(X_train)
df_res = pd.DataFrame(predictions, columns=Y_train.columns)
```

```python
ds_2d_1 = pd.concat([vol for vol in ds_clean_croisiere])
X_train_1 =  ds_2d_1[var]
Y_train_1 = ds_2d_1

model = RandomForestRegressor(n_estimators=10)
model.fit(X_train_1, Y_train_1)

# Étape 3 : On lui demande de "reconstruire" les données
predictions_1 = model.predict(X_train_1)
df_res_1 = pd.DataFrame(predictions_1, columns=Y_train_1.columns)
```
