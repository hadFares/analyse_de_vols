```python
# Import des librairies
from sklearn.feature_selection import mutual_info_regression
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
from pathlib import Path
```

```python
# On place dans le tableu AIRCRAFT_FILES les données extraites par le précédent scrypte.
try:
    parent = Path(__file__).resolve().parent.parent
except NameError:
    parent = Path.cwd().parent

AIRCRAFT_FILES = [
    parent / "aircraft_1_decollage.h5",
    parent / "aircraft_2_decollage.h5",
    parent / "aircraft_3_decollage.h5",
]
```

```python
dfs = [] # On initialise la liste qui va contenir les données à analyser
store = pd.HDFStore(AIRCRAFT_FILES[1], mode="r") # On place dans store la fichier h5,
# mais en mode lecture simplement : On ne charge pas tout le fichier dans la RAM
# Nature de l'objet :HDF5store de la biblioteque pandas

for key in store.keys() : # store.keys contient la liste de toutes les keys

    df = store[key] # dataframe du fichier hdf5
    n = min(25, len(df))
    dfs.append(df.sample(n=n, random_state=42))

store.close()
df = pd.concat(dfs, ignore_index=True) # dataframe final, concaréné, pour lequel on va pouvoir
#calculer les informations mutuelles

# On sépare la colonne EGT du reste pour pouvoir utiliser mutual_info_regression proprement :
y = df["EGT_1 [DEG C]"]
X = df.drop(columns=["EGT_1 [DEG C]"])
```

```python
# Fonction qui renvoie les informations mutuelles entre une matrice et un vecteur
Influences = mutual_info_regression(X, y, discrete_features='auto', n_neighbors=3, copy=True)
```

```python
mi_series = pd.Series(Influences, index=X.columns).sort_values(ascending=False)

top5 = mi_series.head(5)
bot5 = mi_series.tail(5)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

top5.plot(kind="barh", ax=axes[0], color="steelblue")
axes[0].set_title("Top 5 — Plus influents")
axes[0].set_xlabel("Information mutuelle")
axes[0].invert_yaxis()

bot5.plot(kind="barh", ax=axes[1], color="salmon")
axes[1].set_title("Bottom 5 — Moins influents")
axes[1].set_xlabel("Information mutuelle")
axes[1].invert_yaxis()

plt.suptitle("Influence des paramètres sur EGT_1", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

print(Influences)
```

```python
# Matrice de corrélation
corr_matrix = X.corr()

# Heatmap
plt.figure(figsize=(18, 16))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, vmin=-1, vmax=1)
plt.title("Matrice de corrélation des paramètres", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

# Paires avec corrélation > 0.95
seuil = 0.95
paires = []

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > seuil:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            score = corr_matrix.iloc[i, j]
            paires.append((col1, col2, score))

print(f"{len(paires)} paires corrélées à plus de {seuil} :\n")
for col1, col2, score in paires:
    print(f"  {col1} <--> {col2} : {score:.3f}")
```
